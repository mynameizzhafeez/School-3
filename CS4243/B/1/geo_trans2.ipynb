```python
# packages
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#image file name:
imgname = './CS4243_2023_images_small/cat.bmp'
```

```python
# original image reading and showing 
img = cv2.imread(imgname)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#cv2.namedWindow("Original", cv2.WINDOW_NORMAL)
cv2.imshow('Original',img)

M = img.shape
```

```python
# h is the transform matrix
# transform function is opencv warpAffine
#
h = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,h,(M[1],M[0]))

cv2.imshow('Transformed',dst)
```

### Rotation ...

```python
# h is the rotation matrix
# getRotationMatrix2D would be used to creat the rotation matrix
# the 1st and 2nd parameters are rotation center, 3rd is the rotation degree
# transform function is opencv warpAffine
#

h = cv2.getRotationMatrix2D(((M[1]-1)/2.0,(M[0]-1)/2.0),45,1)
print(h)
dst = cv2.warpAffine(img,h,(M[1],M[0]))
cv2.imshow('Rotated',dst)
```

```python

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
h = cv2.getAffineTransform(pts1,pts2)
print(h)
dst = cv2.warpAffine(img, h, (M[1],M[0]))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python

pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
h = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,h,(300,300))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
